# Word Clouds

In order to see how different pronouns/words were described throughout time, I created word clouds of top descriptors for different subcategories.  This notebook only contains the code associated with outputting the word clouds themselves, but please see full process below:
1. Pull all front page articles from the NYTimes, using the NYtimes API and scrapy
2. Using NMF and TFIDF Vectorizer, create topics for specified time periods.  For my purposes, I used election cycles as a good dividing point.  So the time intervals were (1981 - 1984, 1985-1988, .... 2012-2016).  Prior to creating these topics, using a part of speech tagger, I marked all adjectives and adverbs by adding the string 'markedword' to the end of the word. 
3.  Topics were then broken down into subtopics
4.  Word2Vec in Gensim was used to create separate Word2Vec models for each time period by subtopic.
5.  For each time period, you can now look at associated words, and using the previously specified 'markedword' tag, can look at only descriptors.
6.  Now with a list of descriptors and a cosine similarity to the queried word (both from Word2Vec) I created a wordcloud by weighting each word by its cosine similarity.  For example if I queried president, and Word2Vec outputted diplomatic, friendly, evil, with cosine similarities of .80, .53, and .31 respectively I would put together the following string for the word cloud : 'diplomatic diplomatic diplomatic diplomatic diplomatic diplomatic diplomatic diplomatic friendly friendly friendly friendly friendly evil evil evil'.  In the preceding string, the string contains the word diplomatic 8 times, friendly 5 times, and evil 3 times. 

```python
import pandas as pd
import pickle
from textblob import TextBlob
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def get_all_top_words(topic, nmf_models, intervals):
    '''Aggregate all top words across time intervals'''
    all_words = []
    for interval in intervals:
        nmf = nmf_models[interval][0]
        tfidf = nmf_models[interval][1]
        x = get_top_words(nmf, tfidf, 200, topic, keywords = keywords)
        all_words+=x
    return sorted(list(set(all_words)))

def get_top_words(model, tfidf, n_top_words, topic, keywords):
    '''Get top words from specified model'''
    feature_names = tfidf.get_feature_names()
    keys = keywords[topic]
    topic_words = []
    for topic_idx, topic in enumerate(model.components_):
        curr_topic_words=[]
        curr_topic_words+=[feature_names[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]
        overlap = [word for word in curr_topic_words if word in keys]
        if len(overlap)>4:
            topic_words += curr_topic_words
    return topic_words

def query_word2vec(query, categ):
    '''Query wor2vec returning a list of lists containing'''
    records = []
    for interval in intervals:
        #loop through query for each intervals
        list_of_associations = []
        intensities_orig = []
        intensity_scores = []
        curr_record = []
        try:        
            for w,sim in word2vec_models[interval][categ].most_similar(query,topn=1000):
                if (w[-10:]=="markedword"):
                    list_of_associations.append(w[:-12]) ##returns list of all associated descriptors only
                    intensity_scores.append(int(sim*10)) ##provide a 'weight' for each word, their intensity score * 10
                    intensities_orig.append(sim) #keep a hold of for reference
                else:
                    pass
            weighted_words = " "
            for word, intensity_score in zip(list_of_associations, intensity_scores):
                weighted_words = weighted_words+ (word+" ")*intensity_score # provide a string with words
                                                                            # weighted by their relevance
            total_sent = TextBlob(weighted_words).sentiment[0]  # find a sentiment score for assoc. descriptors
            curr_record.append(query)
            curr_record.append(interval)
            curr_record.append(total_sent)
            curr_record.append(list_of_associations)
            curr_record.append(intensities_orig)
            curr_record.append(weighted_words)
            records.append(curr_record)
        except:
            pass
    return records

keywords = {"domestic": ["budget","tax", "house", "senate" ,"congress", "republicans", "deficit", "senator", \
                         "taxes", 'economic','fiscal', 'congressional', 'package', 'legislation', \
                         'policy', 'representative','vote', 'party', 'independent', 'committee','counsel',\
                        'democrats'],
            "middle_east": ["lebanon","beirut","syrian","lebanese","palestinian","palestine","arab", "syria",\
                         "syrians", "israel","israelis", 'jordan', "arafat", "hussein", "jordan", "middle",\
                           "east","kuwait", "saddam", 'afghan', 'iran', 'iraq','iraqi', 'israeli',\
                           'arabia','gaza','west', 'islam', 'islamic', 'sunni', 'isis', 'military', 'troops'\
                           'secretary','afghan', 'pakistan', 'policy', 'al', 'taliban', 'foreign', \
                            'netanyahu','intelligence'],
           "presidential": [ 'al','bob', 'brown', 'bush', 'campaign', 'candidate', 'cheney',\
                                    'clinton', 'convention', 'cruz', 'debate', 'delegates',\
                                    'democratic', 'democrats', 'dole', 'dukakis', 'edwards', 'election', 'ferraro',\
                                    'gingrich', 'gore', 'governor', 'hillary', 'iowa', 'kerry', 'mccain',\
                                    'mondale', 'nomination', 'obama', 'percent', 'perot', 'perry',\
                                    'polls', 'primary', 'race', 'reagan', 'romney', 'ross', 'rubio', 'sanders',\
                                    'santorum', 'trump','vote', 'voters', 'walter']}
```

```python
with open('../data/pickles/model/nmf/word2vec_politics_subclusters.pkl', 'rb') as f:
    word2vec_models = pickle.load(f)
```

```python
with open('../data/pickles/model/nmf/politics_sub_nmf_5topics.pkl', 'rb') as f:
        nmf_models = pickle.load(f)
```

```python
intervals = [(m, m+3) for m in range(1981,2017,4)]
final_dfs = {}
```

```python
topic = 'middle_east'

queries = get_all_top_words(topic, nmf_models, intervals)

add_queries = ['president', 'israel', 'palestine', 'iran', 'iraq', 'terrorism', 'war', 'saddam','congress',\
                      'america','republicans', 'democrats', 'republican', 'democrat', 'india', 'pakistan', 'china']

records = []
queries += add_queries
queries = list(set(queries))
m = 0
for query in queries:
    m+=1
    record = query_word2vec(query, topic)
    records+=record
    
final_dfs[topic] = pd.DataFrame(records, columns = ['word', 'time period','sentiment' ,'associated words', 'sorted_sent_scores','weighted words'])
final_dfs[topic] = final_dfs[topic].drop_duplicates(['word','time period','sentiment'])
final_dfs[topic].sort_values(['sentiment'])[['word','time period','sentiment', 'associated words']].head(20)
```

```python
word = 'president'
topic='domestic'

df_curr = final_dfs[topic][final_dfs[topic]['word']==word][['word','time period','weighted words']].reset_index()
print word
print "----------------------"

wordclouds = []
for m in range(df_curr.shape[0]):
    wordcloud = WordCloud(collocations=False)
    wordclouds.append((wordcloud.generate(df_curr.loc[m]['weighted words']), df_curr.loc[m]['time period']))
    
figs={}
axs={}
for idx,plot in enumerate(wordclouds):
    figs[idx]=plt.figure()
    axs[idx]=figs[idx].add_subplot(111)
    axs[idx].imshow(plot[0])
    plt.imshow(plot[0], interpolation='bilinear')
    plt.axis("off")
    plt.title(plot[1])
    
```
