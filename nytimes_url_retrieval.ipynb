```python
from urllib import request
from pandas.io.json import json_normalize
import pandas as pd
import json
```

```python
api_key = 'daVhn3qA1tWJ4cp77SIthMGLIjN8G1Xo'

df = pd.DataFrame()
for year in range(1981, 2020):
    for month in range(1,13):
        print(year, month)
        curr_request = f'https://api.nytimes.com/svc/archive/v1/{year}/{month}.json?api-key={api_key}'
        contents = request.urlopen(curr_request).read()
        json_data = json.loads(contents)
        curr_df = json_normalize(json_data['response']['docs'])
        df = df.append(curr_df[curr_df['print_page']=='1']).reset_index(drop = True)
```
